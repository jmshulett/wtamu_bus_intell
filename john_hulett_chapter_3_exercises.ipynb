Exercise 1: Read the 5 CSV files

```python
import numpy as np
import pandas as pd
appledata = pd.read_csv("exercises/aapl.csv")
amzdata = pd.read_csv("exercises/amzn.csv")
fbdata = pd.read_csv("exercises/fb.csv")
googdata = pd.read_csv("exercises/goog.csv")
ntflxdata = pd.read_csv("exercises/nflx.csv")
```

Exercise 2: Covert the Date and Volume columns to timestamp and integer respectively and sort by date and ticker

```python
faangdf = pd.read_csv("faang.csv")
faangdf.loc[:,'date'] = pd.to_datetime(faangdf.date)
```

```python
faangdf.loc[:,'volume'] = pd.to_numeric(faangdf.volume)
```

```python
faangdf.sort_values(by=['date', 'ticker'])
```

Find the 7 Rows in Faang with the lowest volumes

```python
faangdf.nsmallest(n=7, columns='volume')
```

```python
melted_faang = faangdf.melt(id_vars=['date', 'ticker'], value_vars=['open', 'high', 'low', 'close', 'volume'], value_name='amount', var_name='Desc')
```

Exercise 6:  

Part A: Read in the Covid19 cases csv file

```python
coviddf = pd.read_csv("exercises/covid19_cases.csv")
```

Part B: Create a date column

```python
coviddf['date'] = pd.to_datetime(coviddf.dateRep)
coviddf2 = coviddf
```

Part C: Set the date column as the index and sort the index

```python
coviddf.set_index('date', inplace=True)
coviddf.sort_index()
```

Part D: Replace all the occurences of United States with USA and United Kingdom with UK

```python
coviddf.loc[(coviddf.countriesAndTerritories == 'United_Kingdom'),'countriesAndTerritories']='UK'
coviddf.loc[(coviddf.countriesAndTerritories == 'United_States_of_America'),'countriesAndTerritories']='USA'
```

```python
coviddf
```

Part E: Filter the data for certain countries into a dataframe

```python
search_values = ['Argentina', 'Brazil', 'China', 'Colombia', 'Inida', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']
covidcleaneddf = coviddf[coviddf['countriesAndTerritories'].str.contains('|'.join(search_values ))]
covidcleaneddf
```

Part F: Pivot the data

```python
covidcleaneddf.reset_index()
```

```python
covidcleaneddf.reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
```
