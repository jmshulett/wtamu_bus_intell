Exercise 10:

Part A

```python
coviddf = pd.read_csv("covid19_cases.csv")
coviddf['date'] = pd.to_datetime(coviddf.dateRep)
coviddf.set_index('date', inplace=True)
coviddf.sort_index()
coviddf.loc[(coviddf.countriesAndTerritories == 'United_Kingdom'),'countriesAndTerritories']='UK'
coviddf.loc[(coviddf.countriesAndTerritories == 'United_States_of_America'),'countriesAndTerritories']='USA'
```

Part B: Five countries with the most cases, find the day with the largest number of cases

```python
topfive = coviddf.groupby('countriesAndTerritories').cases.sum().nlargest().index
coviddf[coviddf.countriesAndTerritories.isin(topfive)].groupby('countriesAndTerritories').cases.idxmax()
```

Part C:

```python
coviddf.groupby(['countriesAndTerritories', pd.Grouper(freq='1D')]).cases.sum().unstack(0).diff().rolling(7).mean().last('7D')[topfive]
```

Part D:

```python
coviddf.reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').drop(columns='China').fillna(0).apply(lambda x: x[(x >0)].idxmin()).sort_values().rename(lambda x: x.replace(' ', ''))
```

Part E:

```python
coviddf.pivot_table(columns='countriesAndTerritories', values='cases', aggfunc='sum').T.transform('rank', method='max', pct=True).sort_values('cases', ascending=False).rename(lambda x: x.replace('', ''))
```
